```python
import matplotlib.pyplot as plt
import numpy as np
import polars as pl
```

```python
data = np.loadtxt('element_results.txt', skiprows=1)
with open('element_results.txt', 'r') as f:
    header = f.readlines()[0].split()
df = pl.DataFrame(data, schema=header)
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(10, 5))
ax = axes[0]
ax.scatter(df.select('eta1'), df.select('eta2'), c=df.select('N1'))
ax = axes[1]
ax.scatter(df.select('x1'), df.select('x2'), c=df.select('N1'))
fig.tight_layout()
```
